Cabin has lots of missing values (~77% missing).

Age missing in ~20% of the rows.

Embarked has very few missing values (just 2).

```python
# Drop Cabin because it's too empty

df.drop(columns = ['Cabin'], inplace = True)
df
```

```python
# checking for outliers

sns.boxplot(df['Age'])
plt.show()
```

```python
# Fill Age with median as outliers are there

df['Age'].fillna(df['Age'].median(), inplace=True)
df
```

```python
# Fill Embarked with mode

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df
```

```python
# Histograms for numerical features

df['Age'].hist(bins=30, figsize=(9,5))
plt.suptitle('Distribution of Age')
plt.show()
```

Age Mostly between 20 to 40 years, 

```python
# Survival Rate by Sex
sns.barplot(x = "Sex", y = "Survived",  data = df)
plt.title("Survival Rate by Sex")
plt.show()
```

-	Females had a higher survival rate than males.

```python
# Survival Rate by Class
sns.barplot(x = "Pclass", y = "Survived",  data = df)
plt.title("Survival Rate by class")
plt.show()
```

-	1st class passengers had higher survival chances than 3rd class

```python
# Age distribution for survived vs non-survived

sns.histplot(data=df, x='Age', hue='Survived', multiple='stack')
plt.title('Age Distribution by Survival')
plt.show()
```

-	Younger passengers had better survival rates.

```python
plt.scatter( df['Age'],df['Fare'])
plt.show()
```

```python
# Fare distribution by Survival
sns.histplot(data=df, x='Fare', hue='Survived', multiple='stack', bins=40)
plt.title('Fare Distribution by Survival')
plt.show()
```

-	Passengers who paid higher fares had better survival rates.
